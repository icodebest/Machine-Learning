```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
# Read the data.
data = np.asarray(pd.read_csv("decision-tree data.csv", header=None))

# converti np array to dataframe
df = pd.DataFrame(data)
df.columns = ['X1', 'X2', 'y']
df.head()

X= df[['X1', 'X2']]
Y= df['y']
```

```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")

model = DecisionTreeClassifier()
model.fit(X, Y)
```

```python
# make predictions
y_pred = model.predict(X)
print(y_pred)
```

```python
# Calculate the accuracy and assign it to the variable acc.
acc = accuracy_score(Y, y_pred)
print(acc)
```

```python
# plot the decision boundary according to the result
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
plt.figure(figsize=(8, 6))
plt.scatter(X['X1'], X['X2'], c=Y, cmap=ListedColormap(['#FF0000', '#00FF00']))
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Decision Boundary')
```

```python
# define boundries and plot the decision boundary
x_min, x_max = X['X1'].min() - 1, X['X1'].max() + 1
y_min, y_max = X['X2'].min() - 1, X['X2'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(['#FF0000', '#00FF00']))
plt.scatter(X['X1'], X['X2'], c=Y, cmap=ListedColormap(['#FF0000', '#00FF00']))
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Decision Boundary')
plt.show()
```

```python
# Define the boundaries of the grid
x_min, x_max = X['X1'].min() - 1, X['X1'].max() + 1
y_min, y_max = X['X2'].min() - 1, X['X2'].max() + 1

# Create a grid of points with a distance of 0.01 between them
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Predict the class for each point in the grid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

# Plot the decision boundary by assigning a color to each point in the mesh
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(['#FF0000', '#00FF00']))

# Plot the training points
scatter = plt.scatter(X['X1'], X['X2'], c=Y, edgecolor='k', s=20, cmap=ListedColormap(['#FF0000', '#00FF00']))
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Decision Boundary')

# Add a legend
legend1 = plt.legend(*scatter.legend_elements(), loc="upper right", title="Classes")
plt.gca().add_artist(legend1)

plt.show()
```
