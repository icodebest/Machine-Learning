```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.random_projection import GaussianRandomProjection
from sklearn.random_projection import SparseRandomProjection
from sklearn.decomposition import PCA


# Load the digits dataset
digits = load_digits()
X = digits.data
y = digits.target

print(f"Original shape: {X.shape}")

# Apply Gaussian Random Projection
n_components = 2  # Reduce to 2 dimensions for visualization
rp=SparseRandomProjection(n_components=n_components)
X_rp = rp.fit_transform(X)

print(f"Reduced shape: {X_rp.shape}")

# Apply PCA for comparison
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

print(f"Explained variance by PCA components: {pca.explained_variance_ratio_}")

# Plot Random Projection results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_rp[:, 0], X_rp[:, 1], c=y, edgecolor='k', cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.title("Random Projection")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.colorbar()

# Plot PCA results
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolor='k', cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.title("PCA")
plt.xlabel(f"Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}% variance)")
plt.ylabel(f"Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}% variance)")
plt.colorbar()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.random_projection import SparseRandomProjection
from sklearn.decomposition import PCA
```

```python
data=load_breast_cancer()
X=data.data
y=data.target

print(f"Original shape: {X.shape}")

# Apply Random Projection
n_components = 2
rp = SparseRandomProjection(n_components=n_components)
X_rp = rp.fit_transform(X)

print(f"Reduced shape: {X_rp.shape}")

# Apply PCA for comparison
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

print(f"Explained variance by PCA components: {pca.explained_variance_ratio_}")

# Plot Random Projection results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_rp[:, 0], X_rp[:, 1], c=y, edgecolor='k', cmap=plt.cm.get_cmap('coolwarm', 2))
plt.title("Random Projection")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.colorbar()

# Plot PCA results
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolor='k', cmap=plt.cm.get_cmap('coolwarm', 2))
plt.title("PCA")
plt.xlabel(f"Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}% variance)")
plt.ylabel(f"Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}% variance)")
plt.colorbar()

plt.show()

```
